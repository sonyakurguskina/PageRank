```python
req = Request("https://datalaboratory.one/")
html_page = urlopen(req)
```

```python
soup = BeautifulSoup(html_page, "lxml")
```

```python
data = {}
data['https://datalaboratory.one/'] = {}
data['https://datalaboratory.one/']['links'] = []
for link in soup.findAll('a'):
    string = link.get('href')
    try:
        req = Request(string)
        html_page = urlopen(req)
        soup = BeautifulSoup(html_page, "lxml")
        data['https://datalaboratory.one/']['links'].append(string)
    except:
        pass
```

```python
for i in range(len(data['https://datalaboratory.one/']['links'])):
    try:
        mainpage = data['https://datalaboratory.one/']['links'][i]
        data[mainpage] = {}
        data[mainpage]['links'] = []
        req = Request(mainpage)
        html_page = urlopen(req)
        soup = BeautifulSoup(html_page, "lxml")
        for link in soup.findAll('a'):
            string = link.get('href')
            try:
                req = Request(string)
                html_page = urlopen(req)
                soup = BeautifulSoup(html_page, "lxml")
                data[mainpage]['links'].append(string)
            except:
                pass
    except:
        pass           
```

```python
print(data)
```

```python
uni = []
for i in data:
    uni = uni + data[i]['links']
unique_links = list(set(uni))
```

```python
uniq = {}
count = 0
uniq[count] = {}
for i in range(len(unique_links)):
    uniq[count] = unique_links[i]
    count += 1
    
```

```python
uniq
```

```python
def get_key(d, value):
    for k, v in d.items():
        if v == value:
            return k
```

```python
matrix = np.zeros((11,650))
```

```python
count = 0
for i in data:
    if len(data[i]['links']) != 0:
        for j in range(len(data[i]['links'])):
            matrix[count][get_key(uniq, data[i]['links'][j])] = 1/len(data[i]['links'])
        count += 1    
    else:
        for f in range(len(uniq)):
            matrix[count][f] = 1/len(uniq)
        count += 1        
```
